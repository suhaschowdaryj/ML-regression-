```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

dataset = pd.read_csv('C:\data\\udemy\Mall_Customers.csv')

print(dataset.head())

X = dataset.iloc[:,3:5].values

#dendrogram
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward' ))
plt.title('Dendogram')
plt.xlabel('customers')
plt.ylabel('eucledian distances')
plt.rcParams['figure.dpi'] = 150
plt.legend()
plt.show()


#agglomerative clustering

from sklearn.cluster import AgglomerativeClustering

hierarchial_cluster = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_predict = hierarchial_cluster.fit_predict(X)

#cluster plot
for i in range(0,6):
    plt.scatter(X[y_predict == i, 0], X[y_predict == i, 1], s = 10, label = i)
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.title('agglomerative clustering(HC)')
plt.rcParams['figure.dpi'] = 150
plt.legend()
plt.show()
```
